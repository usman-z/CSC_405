# Employment Dataset
The employment dataset provides the level of employment and the earning potential by county.

## Task 2

### Lets examine the Employment dataset

```python
import pandas as pd
df = pd.read_excel('allhlcn231.xlsx')
df.head()
```

The Employment dataset for County Quarterly consists of 21 columns total, 16 of which are visible and remaining 5 are hidden.

The 21 columns are as follows:  
- Area Code: FIPS code, a unique identifier for an Area
- St: State FIPS code, a unique identifier for a State
- Cnty: County FIPS code, a unique identifier for a County
- Own: Ownership code
- NAICS: Industry code (SuperSector)
- Year: 4-digit year
- Qtr: 1-character quarter (always A for annual)
- Area Type: Category of the given area. (State, County, Nation, etc.…)
- St Name: Multi-character State name
- Area: Area title associated with the area's FIPS code
- Ownership: Ownership title associated with the ownership code
- Industry: Industry title associated with the industry code
- Status Code: Status code, or disclosure code ('N' for not disclosed)
- Establishment Count: Quarterly establishment counts for a given quarter
- Month 1 Employment: Employment level for the first month of the quarter
- Month 2 Employment: Employment level for the second month of the quarter
- Month 3 Employment: Employment level for the third month of the quarter
- Total Quarterly Wages: Total quarterly wage level for the given quarter
- Average Weekly Wage: Average weekly wage based on the 12-monthly employment levels and total annual wage levels
- Employment Location: Quotient Relative to U.S (Employment Location Quotient Relative to U.S)
- Total Wage Location: Quotient Relative to U.S (Total Wage Location Quotient Relative to U.S.)
