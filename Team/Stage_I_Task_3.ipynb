## Read in the COVID-19 data (cases, deaths, and population) using pandas

```python
import pandas as pd

cases = pd.read_csv("covid_confirmed_usafacts.csv")
deaths = pd.read_csv("covid_deaths_usafacts.csv")
population = pd.read_csv("covid_county_population_usafacts.csv")
```

## Display the dataframes

### Cases dataframe

```python
cases.head()
```

### Deaths dataframe

```python
deaths.head()
```

### County Population dataframe

```python
population.head()
```

## Merge Cases, Deaths, and Population to create a super COVID-19 datafame

```python
cases_deaths_df = pd.merge(cases, deaths, on='countyFIPS', suffixes=('_cases', '_deaths'))
covid_df = pd.merge(cases_deaths_df, population, on='countyFIPS')
covid_df.head()
```

## Export it to a CSV format

```python
covid_df.to_csv("Super COVID19 Data.csv", index=False)
```
