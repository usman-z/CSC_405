# Compare the data against other states (compare against 5 states). Normalize by population, use a normalization factor which is able to identify cases and deaths


### Raw data

```python
## all confirmed cases
cases = mb.read_csv("../../Team/covid_confirmed_usafacts.csv")
cases.head()
```

```python
# all the deaths reported
deaths = mb.read_csv("../../Team/covid_deaths_usafacts.csv")
deaths.head()
```

```python
county_population = mb.read_csv("../../Team/covid_county_population_usafacts.csv")
county_population.head()
```

```python
# Define the date range to select/ We want to look at July-December
start_date = '2020-07-01'
end_date = '2020-12-31'

# Define the list of additional columns
additional_columns = ["countyFIPS", "County Name", "State", "StateFIPS"]

# Use the loc method to select the desired columns and date range
selected_cases = cases.loc[:, additional_columns + list(cases.columns[(cases.columns >= start_date) & (cases.columns <= end_date)])]
selected_deaths = deaths.loc[:, additional_columns + list(deaths.columns[(deaths.columns >= start_date) & (deaths.columns <= end_date)])]
```

### Now we look at all the cases/ deaths for NY

```python
# Reported cases
ny_cases = selected_cases.loc[selected_cases['State'] == 'NY']
ny_cases.head()
```

```python
# Reported deaths
ny_deaths = selected_deaths[selected_deaths["State"] == "NY"]
ny_deaths.head()
```

### Since there is alot of data we create array of cases and deaths on a specific day from each week of NY cases / deaths

```python
# July 1, 2020 is a Wednesday
count = 2

weekly_cases_ny = [] # where Wednesday from each week is chosen
weekly_deaths_ny = []


for col in ny_cases.columns:
    count = count + 1 
    if count % 7 == 0: 
        weekly_cases_ny.append(ny_cases[col].sum())
        
for col in ny_deaths.columns:
    count = count + 1
    if count % 7 == 0:
        weekly_deaths_ny.append(ny_deaths[col].sum())
        
weekly_cases_ny_data = mb.Series(weekly_cases_ny)
weekly_deaths_ny_data = mb.Series(weekly_deaths_ny)
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter(x=weekly_cases_ny_data.index, y=weekly_cases_ny_data, mode='lines', name='New York'))

fig.update_layout(
    title='Weekly Cases in NY',
    xaxis=dict(title='Each Week'),
    yaxis=dict(title='# of Cases'),
    showlegend=True,
    width=1000,
    height=500
)

fig.write_image("ny_cases.png")
Image(filename="ny_cases.png")
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter(x=weekly_deaths_ny_data.index, y=weekly_deaths_ny_data,
                         mode='lines', name='New York'))

fig.update_layout(
    title='Weekly Deaths in NY',
    xaxis=dict(title='Each Week'),
    yaxis=dict(title='# of Deaths'),
    showlegend=True,
    width=1000,
    height=500
)


fig.write_image("ny_deaths.png")
Image(filename="ny_deaths.png")
```

### mean, mode, and median for cases in NY

```python

print("The Average of Cases across Ny from July December 2020: ",round(weekly_cases_ny_data.mean(), 2))

print("The Mode of Cases across Ny from July to December 2020:",weekly_cases_ny_data.mode().values[0])

print("The Median of Cases across Ny from July to December 2020:",int(weekly_cases_ny_data.median()))
```

### mean, mode, and median for deaths in NY

```python

print("The Average of Cases across Ny from July December 2020: ",round(weekly_deaths_ny_data.mean(), 2))

print("The Mode of Cases across Ny from July to December 2020:",weekly_deaths_ny_data.mode().values[0])

print("The Median of Cases across Ny from July to December 2020:",int(weekly_deaths_ny_data.median()))
```

```python
def calculate_weekly_cases(data, county_name):
    count = 2
    weekly_cases = []

    # Filter the data for the specific county
    county_data = data[data["County Name"] == county_name]
    
    for col in county_data.columns:
        count = count + 1
        if count % 7 == 0:
            weekly_cases.append(county_data[col].sum())

    return mb.Series(weekly_cases)
```
